```python
def selection_sort_corrigido(arr):
    n = len(arr)
    print(f"\n--- Início da Ordenação Selection Sort ---")
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
        print(f"Iteração i={i}: min_idx encontrado = {min_idx}. Lista após potencial troca: {arr}")
    print(f"--- Fim da Ordenação ---")
    return arr
```

```python
lista_original = [random.randint(1, 50) for _ in range(7)]
lista_para_ordenar = lista_original.copy()

print(f"Lista Original Aleatória (7 elementos): {lista_original}")

lista_ordenada = selection_sort_corrigido(lista_para_ordenar)

print(f"\nLista Ordenada Final: {lista_ordenada}")
```
